```python
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.io as pio
from   plotly.subplots import make_subplots
```

```python
df = pd.read_csv("Egypt_Houses_Price.csv")
```

```python
df
```

```python
df.info()
```

## Cleaning

```python
df.replace("Unknown", np.nan , inplace =True)
df.replace("Unknown ", np.nan , inplace =True)
```

```python
df.dropna(inplace = True)
```

```python
df = df.convert_dtypes()
```

```python
df["Price"] = df.Price.astype("int32")
df["Bedrooms"] = df.Bedrooms.astype("float32")
df["Bathrooms"] = df.Bathrooms.astype("float32")
df["Area"] = df.Area.astype("float32")
```

```python
df = df.reset_index()
df.drop("index" ,inplace =True ,axis =1)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
index = df[df.City == "(View phone number)"].index
df.drop(index,axis =0 ,inplace =True)
df.shape
```

```python
def price_range(x):
    if(x <= 1000000):
          return 1
    if(x <= 2000000):
           return 2
    if(x <= 3000000):
           return 3
    if(x <= 5000000):
           return 5
    if(x <= 8000000):
           return 8
    if(x <= 10000000):
           return 10
    else:
           return 11
def area_range(x):
    if (x >= 100.0) & (x<= 160.0):
        return "100 to 160"
    elif (x > 160.0) & (x <= 200.0):
        return "160 to 200"
    elif (x > 200.0) & (x <= 250.0):
        return "200 to 250"
    elif (x > 250.0) & (x <= 400.0):
        return "250 to 400"
        
```

```python
df1 = df["Price"].reset_index()
```

```python
df1["Price"] = df1["Price"].apply(price_range)
```

```python
df1
```

```python
price_area = df[["Price","Area"]].reset_index()
price_area["Average Price"] = price_area["Price"].apply(price_range)
price_area["Area"] = price_area["Area"].apply(area_range)
price_area1 = price_area.groupby("Area")["Average Price"].mean()
```

```python
px.bar(data_frame=price_area1 , y = "Average Price",color_discrete_sequence=px.colors.qualitative.Dark2 )
```

```python
px.histogram(data_frame = df1 ,x="Price",nbins=12)
```

* as area inc the price goes up

## price and floor

```python
price_floor = df.groupby('Level').Price.mean().reset_index(name='Price')
price_floor.sort_values("Level",ascending =True)


sns.barplot(data=price_floor,x="Level",y="Price")
plt.title("Price of houses according to the floor level")
plt.xticks(rotation=45)
plt.show()
```
